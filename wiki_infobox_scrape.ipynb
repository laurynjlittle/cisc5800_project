```python
class WikiCleaning():
    
    @classmethod
    def remove_punc(cls, s):

        punc_ = '!"#%&’\'*+,/;<=>?@[\]^`{|}~'

        s = s.translate(str.maketrans('', '', punc_))

        ## remove period standalone or at end of string
        s = re.sub(r'\b[.]{3}',' ',s).strip()
        s = re.sub(r'\b[.](?=\W)','',s)
        s = re.sub(r'\b[.]$','',s)

        return s

    @classmethod
    def remove_text_in_brackets(cls, s):
        s = re.sub(r'\([^)]*\)', '', s).strip()
        s = re.sub(r'\[.*\]','', s).strip()

        s = re.sub(r'\{[^)]*\}', '', s).strip()

        ## remove '(' standalone
    #     s = re.sub(r'[({].*', '', s).strip()
        s = re.sub(r'\b[(](?=\w)',' ',s).strip()
        return s

    @classmethod
    def strip_space(cls,s):
        ## remove any extra space
        s = s.split(" ")
        s = " ".join(n.strip() for n in s if n != "")
        return s

    @classmethod
    def converter(cls, x):
        if 'M' in x:
            return f"{(float(x.strip('M'))*1000000)}"
        elif 'B' in x:
            return f"{(float(x.strip('B'))*1000000000)}"
        else:
            return f"{(float(x)*1)}"
        
        
    @classmethod
    def clean_wiki_budget(cls, df):
        ### clean budget string so it can be converted to float

        ## reformat html text, remove space
        df['budget_cleaned'] = df['budget'].dropna().apply(lambda x: html.unescape(x))
        df['budget_cleaned'] = df['budget_cleaned'].dropna().map(cls.strip_space)

        ## remove text in brackets and punctuation
        df['budget_cleaned'] = df['budget_cleaned'].dropna().map(cls.remove_punc)

        ## filter for budgets reported with '$' to ignore non-USD currency
        df[['budget_cleaned', 'budget_amount']] = pd.Series(df['budget_cleaned']).str.extract(r'([$]\d{1,3}[.,]?\d{0,3}?\S{0,5})\W?(million|billion)?')

        
        if len(df.loc[df['budget'].str.contains(r'US|USD|United States',na=False)]) > 0:
        
            ## extract budget if preceeded by USD/US/etc.
            df[['budget_cleaned2', 'budget_amount2']]= pd.Series(df.loc[df['budget'].str.contains(r'US|USD|United States',na=False)]['budget']).astype(str).str.extract(
                r'([$]\d{1,3}[.]?\d{0,3}?)\W+(million|billion)')

            ## extract budget if preceeded by USD/US/etc.
            df[['budget_cleaned3', 'budget_amount3']]= pd.Series(df.loc[df['budget'].str.contains(r'US|USD|United States',na=False)]['budget']).astype(str).str.extract(
                r'(?<=USD)(\d{1,3}[.]?\d{0,3}?)\W+(million|billion)')

        else:
            df[['budget_cleaned2', 'budget_amount2', 'budget_cleaned3', 'budget_amount3']] = np.nan
            
        ## combine the extracted budgets into single column

        df.loc[(df['budget_cleaned2']==0), 'budget_cleaned2'] = df['budget_cleaned3']
        df.loc[(df['budget_amount2']==0), 'budget_amount2'] = df['budget_amount3']

        df.loc[(df['budget_cleaned']==0), 'budget_cleaned'] = df['budget_cleaned2']
        df.loc[(df['budget_amount']==0), 'budget_amount'] = df['budget_amount2']

        df.drop(['budget_cleaned2', 'budget_cleaned3', 'budget_amount2', 'budget_amount3'], axis=1, inplace=True)

        ## replace 0 with Nan
        df.loc[df['budget_amount'] == 0, 'budget_amount'] = np.nan
        df.loc[df['budget_cleaned'] == 0, 'budget_cleaned'] = np.nan

        ## for any still null...
        df['budget_nulls'] = df[(df['budget_amount'].isna())]['budget'].dropna().map(cls.remove_text_in_brackets).map(cls.strip_space)
        df['budget_nulls'] = pd.Series(df['budget_nulls'].dropna()).str.extract(r'(million|billion)')
        df.loc[(df['budget_cleaned'].notnull()) & (df['budget_amount'].isna()) & 
               (df['budget_nulls'] != 0), 'budget_amount'] = df['budget_nulls']
        df.drop(['budget_nulls'], axis=1, inplace=True)

        ## remove '$'
        df['budget_cleaned']=df['budget_cleaned'].dropna().apply(lambda x: re.sub('[$]','',str(x))).map(cls.strip_space)
        df['budget_amount']=df['budget_amount'].dropna().apply(lambda x: pd.Series(re.sub(r'million','M',str(x)).strip()))

        ## replace '-' variations with the same symbol
        df['budget_cleaned']=df['budget_cleaned'].dropna().apply(lambda x: re.sub('–|‒|—|−','-',x))

        ## split '-' and use the larger number in range
        df['budget_cleaned'] = df['budget_cleaned'].dropna().apply(lambda x: x.split('-')[-1:][0])
        df.loc[~df['budget_amount'].isna(), 'budget_cleaned'] = df['budget_cleaned'] + df['budget_amount']
        
        df['budget_cleaned'] = df['budget_cleaned'].dropna().apply(cls.converter)
        df['budget_cleaned'] = df['budget_cleaned'].astype(float)

        return df
    
    @classmethod
    def clean_wiki_release(cls, df):
        
        ## reformat html text, remove space
#         df['released_cleaned'] = df['released'].apply(lambda x: html.unescape(x))
        df['released_cleaned'] = df['released'].dropna().map(cls.strip_space)
        
        ## extract if preceeded by US
        df['released_cleaned']= pd.Series(df.loc[df['released_cleaned'].str.contains(r'United States', na=False)]['released_cleaned']).astype(str).str.extract(
            r'(\d{4}[|]\d{1,2}[|]\d{1,2})[|]?(?>\S{0,2}United States\S{0,2}|United Kingdom\S{1,6}United States)')

        df['released_cleaned2']= pd.Series(df.loc[df['released'].str.contains(r'United States',na=False)]['released']).astype(str).str.extract(
            r'(\d{4}[|]\d{1,2}[|]\d{1,2})[|]\S+(?>United States)')

        df.loc[df['released_cleaned'] == 0, 'released_cleaned'] = df['released_cleaned2']
        
        ## if released cleand is null, use only date available in field
        df.loc[df['released_cleaned'].isna(), 'released_cleaned'] = pd.Series(df['released'].dropna()).str.extract(
            r'(\d{4}[|]\d{1,2}[|]\d{1,2})')
        
        df.loc[df['released_cleaned'].isna(), 'released_cleaned'] = pd.to_datetime(df['released'], errors='coerce')
        
        ## remove punctuation
        df['released_cleaned'] = df['released_cleaned'].astype(str).apply(lambda x: re.sub('[|]','-',x))
        df.loc[df['released_cleaned'] == '0', 'released_cleaned'] = np.nan
        df['released_cleaned'] = pd.to_datetime(df['released_cleaned'])
        
        return df
```
